```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
fn=glob.glob("*.txt")
print(fn)
```

```python
df={}
df["date"]=[]
df["T"]=[]
df["H"]=[]
df["J"]=[]
for i in fn:
    data=open(i).read().split("\n")
    print(len(data))
    ic=0
    for j in data:
        if ic==0:
            r=j.split(",")
            print(r[0],r[1],r[5],r[6])
        else:
            r=j.split(",")
            df["date"].append(r[0])
            df["T"].append(float(r[1]))
            df["H"].append(float(r[5]))
            df["J"].append(float(r[6]))
        ic+=1
```

```python
try:
    df=pd.DataFrame.from_dict(df)
except:
    pass
display(df.head())
df.to_csv("vreme.csv")
```

```python
df_houry={}
df_houry["date"]=[]
df_houry["time"]=[]
df_houry["T"]=[]
df_houry["J"]=[]

df["day"]=df["date"].apply(lambda x:x.split(" ")[0])
df["HH"]=df["date"].apply(lambda x:x.split(" ")[1].split(":")[0]+":00")

dg=df.groupby("day")
for i,ii in dg:
    dh=ii.groupby("HH")
    for j,jj in dh:
        df_houry["date"].append(i)
        df_houry["time"].append(j)
        df_houry["T"].append(jj["T"].mean())
        df_houry["J"].append(jj["J"].mean())
df_houry=pd.DataFrame.from_dict(df_houry)
display(df_houry)
df_houry.to_csv("Vreme_urni.csv",index=False)
```
