```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random
import warnings
warnings.filterwarnings('ignore')
```

Ustvarimo funkcijo Morita, s katero bomo ustvarili različne mrežne topologije. Pri tem so pomembni parametri:<br>
<ul>
<li>$N$ - Število vozlišč</li>
<li>$k_{avg}$ - Povprečna povezanost</li>
<li>$pos$ - lega posamznih vozlišč</li>
<li>$Fitness$ - atraktivnost vozlišča</li>
<li>$delta$ - obtežitev razdalje glede na atraktivnost</li>
</ul>

```python
def Morita(N,kavg,delta,plot):
    #1) Določimo lege (2D)
    pos=np.zeros([N,2],float)
    for i in range(N):
        pos[i,0]=np.random.rand() #Lega v smeri x za vozlišče i
        pos[i,1]=np.random.rand() #Lega v smeri y za vozlišče i
    
    #2) Poznati moramo razdalje med vozlišči !!(N*(N-1))/2 !!
    I=np.zeros([N,N],float)
    for i in range(N):
        xi,yi = pos[i]
        for j in range(i+1,N):
            xj,yj = pos[j]
            dx=(xi-xj)**2
            dy=(yi-yj)**2
            I[i,j]=dx+dy
            I[j,i]=I[i,j]
    I=np.power(I,delta/2.0)
    
    #3) Atraktivnost
    f=np.zeros([N],float)
    n=1.0/(1.0-float(2.5))
    for i in range(N):
        y=(i+1)/float(N)
        f[i]=np.power(y,n)
    
    #4) optimizacija delovanja
    for i in range(N):
        for j in range(i+1,N):
            I[i,j]=f[i]*f[j]/I[i,j]
            I[j,i]=I[i,j]
            
    #5) Ustvarimo mrežo
    THETA=100.0
    k=0.0 #Trenutna povprečna povezanost
    x=[]
    y=[]
    ic=0
    while abs(k-kavg)>0.1:
        #Ustvarim si prazen graf (meža)
        G=nx.Graph()
        
        #Graf poselim z vozlišči
        for i in range(N):
            G.add_node(i)
        
        #Graf: ustvarimo povezave med vozlišči
        for i in range(N):
            for j in range(i+1,N):
                if I[i,j]>THETA:
                    G.add_edge(i,j)
        
        #Preverimo ali mreža ustreza našim kriterijem
        k=np.mean([G.degree(i) for i in G])
        if k>kavg: #Sprejmemo preveč povezav, torej je prag prenizek
            THETA=THETA*1.01 #Za malo povečamo theta
        if k<kavg: #Sprejmemo premelo povezav, torej je prag previsok
            THETA=THETA*0.98 #Theto malo zmanjšamo
        x.append(ic)
        y.append(k)
        ic+=1.0
    if plot:
        plt.plot(x,y,marker='o')
        plt.hlines(kavg,0,ic,color='red')
        plt.show()
    return [G,pos]
```

# Ustvarimo mrežo

```python
N=100
kavg=6.0
delta=1.1

G,pos = Morita(N,kavg,delta,True)
```

# Vizualizacija mreže

```python
fig, ax =plt.subplots(2,5,figsize=(15,6))

d=[0.1,1.1,2.1,3.1,4.1] #Določimo delte

for iax,delta in enumerate(d):
    G,pos = Morita(N,kavg,delta,False)
    ax[0,iax].set_title(r'$\delta$ = %.1f'%(delta))
    nx.draw(G,
            pos=pos,
            node_size=[5*G.degree(i)+5 for i in G],
            node_color='red',
            ax=ax[0,iax])
    ax[1,iax].hist([G.degree(i) for i in G],width=0.8)
    ax[1,iax].set_xlabel('k')
    ax[1,iax].set_ylabel('N(k)')
    print(iax)
plt.show()
```

<h1>Domača naloga:</h1>
<ul>
    <li>Določite kolikšna je povprečna dolžina povezav zgoraj generiranih mrež. </li>
    <li>Kako se obnašajo mreže (kolikšna je povprečna povezanost mreže), če vozlišča naključno odstranjujemo (G.remove_node(i))</li>
    <li>Kako se obnašajo mreže (kolikšna je povprečna povezanost mreže), če vozlišča odstranjujemo ciljno-. To pomeni, da zmeraj odstranimo vozlišče, ki ima največjo povezanost v mreži.</li>
    <li>Katera mreža je bolj občutljiva na ciljne napade in katera na naključne.</li>
</ul>
