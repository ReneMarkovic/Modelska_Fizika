```python
import spacy
import nltk
from tqdm.notebook import tqdm
import networkx as nx
import numpy as np
nlp=spacy.load("en_core_web_sm")
```

```python
book=open("GOT.txt",encoding="utf8").read().split("\n")
```

```python
print(len(book))
```

```python
persons={}
```

```python
stop=int(len(book))
for para,text in tqdm(enumerate(book[0:stop]),total=stop):
    text=text.replace("“","")
    text=text.replace("”","")
    #print(text)
    doc=nlp(text)
    for ent in doc.ents:
        if ent.label_=="PERSON":
            if ent.text not in persons:
                persons[ent.text]=[]
            persons[ent.text].append(para)
            #print(para,ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
G=nx.Graph()
for p in persons:
    G.add_node(p)
```

```python
for p1 in G:
    x=np.array(persons[p1])
    for p2 in G:
        y=np.array(persons[p2])
        if p1!=p2:
            xy=np.array([np.min(np.abs(y-xi)) for xi in x])
            wij=len(np.where(xy<2)[0])
            if wij>2:
                print(p1,p2,wij)
                G.add_edge(p1,p2,weight=wij)
```

```python
remove_nodes=[]
for node in G:
    if G.degree(node)<1:
        remove_nodes.append(node)

for node in remove_nodes:
    G.remove_node(node)
    
for node in G:
    print(node,G.degree(node))
```

```python
nx.write_graphml(G,"sociogram.graphml")
```
